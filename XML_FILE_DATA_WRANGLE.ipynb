# Task 1-
10 countries with the lowest infant mortality rates


```python
doc=ET.parse(r'C:\Users\abhij\Desktop\data_wrangling_xml\data\mondial_database.xml')
```

```python
emp_dict={}

for element in doc.iterfind('country'):
    for i in element.iterfind('infant_mortality'):
        emp_dict[element.find('name').text]=[float(element.find('infant_mortality').text)]
        
```

```python
import pandas as pd
from pandas import DataFrame,Series
df=DataFrame.from_dict(emp_dict,orient='index')

df.columns=['Infant Mortality Rate']

df.sort_values('Infant Mortality Rate').head(10)

```

# Task 2 :10 cities with the largest population

```python
emp_dict1={}

for element in doc.iterfind('country'):
    for i in element.iterfind('city'):
        for x in i.iterfind('population'):
            emp_dict1[i.find('name').text]=[int(i.find('population').text)]
    
```

# Task 3 : 10 ethnic groups with the largest overall populations (sum of best/latest estimates over all countries)


```python
# First I will give you a general idea of what I'm trying to do here:
#Fisrt extract a dictionary from the xml file containing the Country and Population then convert it into a DataFrame1


#The Second part consist of a list containing Counytry,Ethnic Group and Percentage then convert it into DataFrame2
#Finally the last part is to merge the two dataframes through the common attribute Country
```

```python
#Step 1 To extract the Country and its corresponding population (its latest year)
```

```python
#initiate an empty dict

empty_dict={}

for element in doc.iterfind('country'):
    empty_dict[element.find('name').text]=[int(element.find('population[last()]').text)]
```

```python
# We now have sucessfully extracted a dict containing Country and Population (latest)
empty_dict
```

```python
# Now convert the above dict into a pandas dataframe
df1=DataFrame.from_dict(empty_dict,orient='index')
```

```python
df1.reset_index(drop=False,inplace=True)
```

```python
df1.columns=['Country','Population']
df1.head()
```

```python
############ Objective  1 finished
```

```python
# Step 2 Extract a list containing the Country,Ethnic Group and Percentage
```

```python
eth_list=[]

for element in doc.iterfind('country'):
    c_name=element.find('name').text
    for x in element.iterfind('ethnicgroup'):
        eth_per=x.attrib['percentage']
        eth_name=x.text
        eth_list.append([c_name,eth_name,eth_per])
```

```python

eth_list
```

```python
df2=DataFrame(eth_list,columns=['Country','Ethnic Group','Percentage'])
```

```python
# We have sucessfully extracted the second dataframe which contains Country,Ethnic Group and Percentage

df2.head(5)
```

```python
# Merging df1 and df2 using pandas merge

```

```python
df2=df2.merge(df1[['Country','Population']],on='Country')
```

```python
df2.head()
```

```python
df2['Population Final']=(df2['Percentage'].astype(float))*df2['Population']
```

```python
order_df=df2.groupby('Ethnic Group').sum()
```

```python
order_df.sort_values('Population Final',ascending=False).head(10)
```

# Task 4:name and country of a) longest river, b) largest lake and c) airport at highest elevation

```python
# for country and codes


empty_dict={}

for element in doc.iterfind('country'):
    empty_dict[element.attrib['car_code']]=element.find('name').text
```

```python
empty_dict
```

```python
r_empty_list=[]

for element in doc.iterfind('river'):
    r_name=element.find('name').text
    r_length=element.find('length')
    if r_length!=None:
        r_len=r_length.text
        for x in element.attrib['country'].split():
            new_code=empty_dict[x]
    r_empty_list.append([new_code,x,r_name,float(r_len)])
```

```python
river_label=['Country','Country Code','River','Length']
```

```python
r_df=DataFrame(r_empty_list,columns=river_label)
```

```python
r_df.head(3)
```

```python
r_df.sort_values('Length',ascending=False).head(1)
```

```python
airport_empty_list=[]

for element in doc.iterfind('airport'):
    a_name=element.find('name').text
    a_elevation=element.find('elevation')
    if a_elevation!=None:
        a_elev=a_elevation.text
        for x in element.attrib['country'].split():
            new_code=empty_dict[x]
    airport_empty_list.append([new_code,x,a_name,a_elev])
```

```python
airport_empty_list
```

```python
airport_label=['Country','Country Code','Airport Name','Elevation Point']
```

```python
air_df=DataFrame(airport_empty_list,columns=airport_label)
```

```python
#https://pandas.pydata.org/pandas-docs/stable/generated/pandas.to_numeric.html

air_df1=air_df.apply(pd.to_numeric,errors='ignore')
air_df1.head()
```

```python
air_df1.sort_values('Elevation Point',ascending=False).head(1)
```

```python
empty_lake_list=[]


for element in doc.iterfind('lake'):
    lake_name=element.find('name').text
    lake_area=element.find('area')
    if lake_area!=None:
        lake_area_new=lake_area.text
        for x in element.attrib['country'].split():
            new_code=empty_dict[x]
    empty_lake_list.append([new_code,x,lake_name,lake_area_new])
    
```

```python
lake_df=DataFrame(empty_lake_list,columns=['Country','Country Code','Lake Name','Area of Lake'])
```

```python
lake_df1=lake_df.apply(pd.to_numeric,errors='ignore')
lake_df1.head(3)
```

```python
lake_df1.sort_values('Area of Lake',ascending=False).head(1)
```
